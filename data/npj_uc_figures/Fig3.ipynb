```python
from countrycrab import cosmetics # this is optional to reproduce the exact graphics of the paper
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Fig. 3a - ITS 3-SAT

### Load data

```python
data_3sat = pd.read_csv('fig3_its_3sat.csv')
```

### Plot data

```python
barwidth = 4
ax = cosmetics.new_fig(2.3,2.3) # this is optional to reproduce the exact graphics of the paper
blue_colors = plt.cm.Blues(np.linspace(0.2,0.8,3))



ax.semilogy(data_3sat['V'], data_3sat['HNN-Q'],'-o' , color = 'tab:purple',alpha=0.8,label ='HNN-Q',markersize = 10)
ax.semilogy(data_3sat['V'], data_3sat['HNN-P'], '-o',color = 'tab:pink',alpha=0.8,label ='HNN-P',markersize = 10)
ax.semilogy(data_3sat['V'], data_3sat['WalkSAT'], '-o',color = 'tab:red',alpha=0.8,label = 'WalkSAT',markersize = 10)
ax.semilogy(data_3sat['V'], data_3sat['MNSAT'], '-s',color = blue_colors[0],label = 'MNSAT',markersize = 10)
ax.semilogy(data_3sat['V'], data_3sat['GNSAT-U'], '-d',color = blue_colors[1],label = 'GNSAT-U',markersize = 10)
ax.semilogy(data_3sat['V'], data_3sat['GNSAT-N'], '-*',color = 'tab:blue',label ='GNSAT-N',markersize = 10)

plt.yscale('log')
plt.ylabel('Median ITS')
plt.xlabel('Number of variables V')
plt.ylim(1e1,1e7)
plt.xlim(0,160)
plt.xticks([0,50,100,150])
plt.tight_layout()
```
