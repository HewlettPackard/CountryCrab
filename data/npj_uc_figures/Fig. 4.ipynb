```python
from countrycrab import cosmetics # this is optional to reproduce the exact graphics of the paper
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Fig. 4a - Energy per cycle 3-SAT

### Load data

```python
e_cycle = pd.read_csv('fig4_energy_per_cycle_3sat.csv')
```

### Plot data

```python
ax = cosmetics.new_fig(2.3,2.3) # this is optional to reproduce the exact graphics of the paper
blue_colors = plt.cm.Blues(np.linspace(0.2,0.8,3))

tmp_df = e_cycle[e_cycle['heuristic']=='mnsat']
ax.plot(tmp_df['N'],tmp_df['energy_mean']*1e12,'-s',color = blue_colors[0],label = 'KLIMA-M',markersize = 10)
tmp_df = e_cycle[e_cycle['heuristic']=='gnsat-u']
ax.plot(tmp_df['N'],tmp_df['energy_mean']*1e12,'-d',color = blue_colors[1],label = 'KLIMA-G-U',markersize = 10)
tmp_df = e_cycle[e_cycle['heuristic']=='gnsat-n']
ax.plot(tmp_df['N'],tmp_df['energy_mean']*1e12,'-*',color = 'tab:blue',label = 'KLIMA-G-N',markersize = 10)
ax.set_xlabel('Number of variables N')
ax.set_ylabel('Energy per cycle [pJ]')
ax.set_ylim(0,150)
#plt.legend()
plt.tight_layout()
```

# Fig. 4c - ITS 3-SAT

### Load data

```python
tts_3sat = pd.read_csv('fig4_tts_3sat.csv')
```

### Plot data

```python
barwidth = 4
ax = cosmetics.new_fig(2.3,2.3) # this is optional to reproduce the exact graphics of the paper
blue_colors = plt.cm.Blues(np.linspace(0.2,0.8,3))

ax.semilogy(tts_3sat['V'], tts_3sat['memHNN-Q'], '-o',color = 'tab:purple',label ='QUBO',markersize = 10)
ax.semilogy(tts_3sat['V'], tts_3sat['memHNN-P'], '-o',color = 'tab:pink',label ='PUBO',markersize = 10)
ax.semilogy(tts_3sat['V'], tts_3sat['KLIMA-M'], '-s',color = blue_colors[0],label = 'KLIMA-M',markersize = 10)
ax.semilogy(tts_3sat['V'], tts_3sat['KLIMA-G-U'],'-d',color = blue_colors[1],label = 'KLIMA-G-U',markersize = 10)
ax.semilogy(tts_3sat['V'], tts_3sat['KLIMA-G-N'],  '-*',color = 'tab:blue',label ='KLIMA-G-N',markersize = 10)
ax.semilogy(tts_3sat['V'], tts_3sat['CPU-WalkSAT'], '-o',color = 'tab:red',alpha=0.8,label = 'CPU-WalkSAT',markersize = 10)

plt.yscale('log')
plt.ylabel('Median TTS [s]')
plt.xlabel('Number of variables V')
plt.ylim(1e-7,1e0)
plt.xlim(0,160)
plt.xticks([0,50,100,150])
plt.tight_layout()
```
